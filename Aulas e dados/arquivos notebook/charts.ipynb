```python
# Carregamento da df de dados

df = pd.read_csv('trees.csv')

df.shape

#dados

df.head()
```

```python


# Criação do histograma, considerando somente o segundo atributo do df de dados e com duas divisões (bins)

# A variável 'h' armazena as faixas de valores de Height

h = np.histogram(df.iloc[:,1], bins = 6) #df.loc[] localiza com base no nome da coluna , o df.iloc[] com base no índice

h




```
